### Least squares

Our goal is to <i>approximate</i> the solution for an overdetermined linear system (more equations (rows) than variables (colums)). 
$$
A x = y.
$$
For an approximation $\hat{x}$, consider the squared-error
$$
\begin{eqnarray}
e &=& ||y - A \hat{x}||^2 \\
  &=& (y - A \hat{x})^t (y - A \hat{x}) \\
  &=& (y^t - \hat{x}^t A^t) (y - A \hat{x}) \\
  &=&  y^t y - A^t y \hat{x} - \hat{x}^t A^t y + \hat{x}^t A^t A \hat{x}.
\end{eqnarray}
$$
In order to minimize the error, take the derivative and set it to zero 
$$
\begin{eqnarray}
\frac{\partial e}{\partial \hat{x}} &=& 0 \\ 
-A^t y - A^t y + 2 A^t A \hat{x}&=& 0  \\
A^t A \hat{x}&=&  A^t y.
\end{eqnarray}
$$
We know that $A^t A$ is a square symmetric matrix. But more importantly, $A^t A$ is <i>positive semi-definite</i>, which means that it is invertible! This fact allows us to solve for $\hat{x}$ 
$$
\begin{eqnarray}
\hat{x} = (A^t A)^{-1} A y^t.
\end{eqnarray}
$$
Implemented below - 

```python
def least_squares(X, y, order=2):    
    A = np.array([X**i for i in range(order+1)]).T 
    x = np.linalg.inv(A.T @ A) @ (A.T @ y)
    return x
```

This can be further simplified with the QR factorization of $A$ (LU decomposition for symmetric matrices). 

><b>QR Decomposition - </b> $A = QR$, where
>1. $Q$ is an orthogonal matrix $\implies Q^t = Q^{-1} \implies Q Q^t = Q^t Q = I$.
>2. $R$ is an upper triangular matrix (may or may not have an inverse).

Substitute $QR$ factorization into solution
$$
\begin{eqnarray}
\hat{x} &=& (A^t A)^{-1} A^t y \\
        &=& ((Q R)^t Q R)^{-1} (Q R)^t y \\
        &=& (R^t Q^t Q R)^{-1} (R^t Q^t) y \\ 
        &=& (R^t R)^{-1} (R^t Q^t) y \\ 
        &=& R^{-1} R^{-t} R^t Q^t y \\ 
        &=& R^{-1} Q^{t} y.
\end{eqnarray}
$$
implemented below- 

```python
# you can also decompose further using QR factorization (you probably want to add some exception handling if R is invertiblef)
def least_squares_qr(X, y, order=2):    
    A    = np.array( [X**i for i in range(order+1)] ).T
    Q, R = np.linalg.qr(A)
    x    = np.linalg.inv(R) @ Q.T @ y
    return x
```

```python
# generate some random data by randomly shifting points on a line 
sol     = least_squares(X, Y, order=2)

plt.scatter(X, Y, s=1)

x = np.linspace(0, 1, 100)
y = sol[1]*x + sol[0]
plt.plot(x, y, c='r')
```

## Condition number

Consider the linear system:
$$
Ax = b.
$$

1. How does $b$ change when you make small changes to $x$? (This is an important question for iterative methods such as conjugate gradient.)
2. The condition number is a numerical measurement of this property.

Consider the following two quantities:
$$
\begin{align*}
a &= \max \frac{||Ax||}{||x||} \\
b &= \min \frac{||Ax||}{||x||}
\end{align*}
$$

Notice how both quantities measure how the vector $x$ is <i>stretched</i> ($a$) or <i>compressed</i> ($b$).

The condition number $\kappa$ is simply the ratio of the max and min norms:
$$
\kappa = \frac{a}{b} = \frac{\max \frac{||Ax||}{||x||}}{\min \frac{||Ax||}{||x||}}.
$$

Notice that we are not specifying the vector $x$. We are simply looking at the highest possible change in magnitude due to the operator $A$ over all possible values of $x$.

> <b> What is the condition number if the matrix $A$ is singular (or close to being singular)?</b>

$A$ is <i>singular</i> $\implies$ there exists a vector $x \neq 0$ such that $Ax = 0$ $\implies \min \frac{||Ax||}{||x||} = 0 \implies \kappa = \infty$.

> <b> What does this imply in terms of the system's solution?</b>

The larger the condition number $\kappa$, the <i>closer</i> the matrix is to being singular, and the harder it is to accurately solve the linear system.
